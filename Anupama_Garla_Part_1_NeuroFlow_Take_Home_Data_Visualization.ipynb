# 2. Data Visualization of Patient Progress on GAD-7

### Challenge: Visualize single patient progress on GAD-7 for provider and patient.

```python
# Isolate single patient data by inputting their patient_id number.
single_patient = GAD_7_df[GAD_7_df["patient_id"]==0]

# Extract Patient ID to label visualization
patient_id = single_patient["patient_id"].values[0]

# Visualization Styling
sns.despine(right = True, top=True)
sns.set(font_scale=1)
sns.set_style("whitegrid")

plt.figure(figsize=(20,10))

# Visualization Content
chart = sns.lineplot(
    x='date', 
    y='score',
    data=single_patient,
    linewidth = 2.5
    ).set(title = 'GAD-7 Results Time Series, Patient ID: %i' %patient_id, 
                             xlabel='Date', ylabel='Score', yticks = [0,5,10,15,20,25])
# control x and y limits
plt.ylim(0, 25)
```
